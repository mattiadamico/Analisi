```python
from ebooklib import epub
import ebooklib
from bs4 import BeautifulSoup
```

```python
import tiktoken

```

```python
def estrai_testo(filepath):
    risultato = ''
    doc = epub.read_epub(filepath)
    for item in doc.get_items_of_type(ebooklib.ITEM_DOCUMENT):
        soup = BeautifulSoup(item.get_body_content().decode(),'html.parser')
        text = [paragrafo.get_text().strip() for paragrafo in soup.find_all('p')]
        text = ' '.join(text)
        risultato += text
    return risultato

def estrai_testo_verso_lista(filepath):
    risultato = []
    doc = epub.read_epub(filepath)
    for item in doc.get_items_of_type(ebooklib.ITEM_DOCUMENT):
        soup = BeautifulSoup(item.get_body_content().decode(),'html.parser')
        risultato.extend([paragrafo.get_text().strip() for paragrafo in soup.find_all('p')])
    return risultato
    
filepath = r"EPUB\leggi 2020\53613.epub"
#print(estrai_testo(filepath))
```

```python
def num_tokens_from_string(string: str, encoding_name = "cl100k_base") -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
```
